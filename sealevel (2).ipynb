```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
 # Create scatter plot
x = df['Year']
y = df['CSIRO Adjusted Sea Level']


plt.scatter(x,y)
plt.title("scatter plot for the rise in sea level over the years")
plt.ylabel("CSIRO Adjusted Sea Level")
plt.xlabel("Year")
```

```python
#finding linear regression
result = linregress(x, y)
slope = result.slope
intercept = result.intercept
y_max = slope*x + intercept #y = mx+c

plt.scatter(x,y,color="green")
plt.plot(x,y_max,color="red")
plt.xlabel("Year")
plt.ylabel("Rise in Sea level")
plt.title("Linear Regression compared to scatter plot")
plt.show()




```

```python
#second regression
# second line of best fit
df_prediction = df.loc[df['Year'] >= 2000]
x_pred = df_prediction['Year']
y_pred = df_prediction['CSIRO Adjusted Sea Level']

# get new slope + intercept
result2= linregress(x_pred, y_pred)
slope= result2.slope 
intercept = result2.intercept
stderr= result2.stderr

x_pred2 = pd.Series([z for z in range(2000, 2051)])
y_pred2 = slope*x_pred2 + intercept
plt.plot(x_pred2, y_pred2, 'blue')
# Add labels and title
plt.title('Prediction of Rise in Sea Level from 2000-2050')
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Sea Level (inches)', fontsize = 15)
#Save plot
plt.savefig('sea_level_plot.png')
    
```
